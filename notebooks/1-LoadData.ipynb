```python
from ipumspy import readers
import pandas as pd
```

```python
ddi1 = readers.read_ipums_ddi("usa_00011.xml")
df1  = readers.read_microdata(ddi1, "usa_00011.dat.gz")

df1.head()
```

```python
ddi2 = readers.read_ipums_ddi("usa_00012.xml")
df2  = readers.read_microdata(ddi2, "usa_00012.dat.gz")

df2.head()
```

```python
df3 = pd.concat([df2, df1], ignore_index = True)
```

```python
df3.to_parquet("acs.parquet", index=False)
```
