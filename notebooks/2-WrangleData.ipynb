```python
df = df.rename(columns = {"YEAR": "Year",
                          "STATEFIP": "State FIP Code",
                          "PERWT": "Individuals Represented",
                          "SEX": "Sex",
                          "AGE": "Age",
                          "RACE": "Race",
                          "RACED": "Race (Detailed)",
                          "HISPAN": "Hispanic",
                          "HISPAND": "Hispanic (Detailed)",
                          "EDUC": "Education Level",
                          "EDUCD": "Education Level (Detailed)",
                          "DEGFIELD": "Major",
                          "DEGFIELDD": "Major (Detailed)",
                          "EMPSTAT": "Employment Status",
                          "EMPSTATD": "Employment Status (Detailed)",
                          "WKSWORK1": "Weeks Worked",
                          "UHRSWORK": "Hours Worked",
                          "INCWAGE": "Income (Nominal)"})
```

# **Filter Data**
---

```python
# Restrict to working-age individuals
df = df[(df['Age'] >= 25) & (df['Age'] <= 64)]
```

```python
# Restrict to employed individuals
df = df[df['Employment Status'] == 1]
```

```python
# Restrict to individuals with positive incomes below the confidentiality cap
df = df[(df['Income (Nominal)'] > 0) & (df['Income (Nominal)'] < 999999)]
```

```python
# Restrict to full-time individuals 
df = df[(df['Weeks Worked'] >= 40) & (df['Hours Worked'] >= 35)]
```

```python
# Remove years where hours worked don't exist
years_with_hours = (
    df.groupby("Year")["Hours Worked"]
      .apply(lambda x: x.notna().any())
)

years_with_hours = years_with_hours[years_with_hours].index

df = df[df["Year"].isin(years_with_hours)]
```

# **Map Data**
---

```python
df["Sex"] = df["Sex"].map({1: "Male", 2: "Female"})
```

```python
def clean_race(row):
    if row['Hispanic'] in [1,2,3,4]:
        return "Hispanic"
    
    if row['Race'] == 1:
        return "White"
    elif row['Race'] == 2:
        return "Black"
    elif row['Race'] in [4,5,6]:
        return "Asian"
    else:
        return "Other"

df['Demographic'] = df.apply(clean_race, axis=1)
```

```python
major_map = {
    0: np.nan,
    11: "Agriculture",
    13: "Environment and Natural Resources",
    14: "Architecture",
    15: "Area/Ethnic/Civilization Studies",
    19: "Communications",
    20: "Communications Technologies",
    21: "Computer and Information Sciences",
    22: "Culinary and Cosmetology",
    23: "Education Administration and Teaching",
    24: "Engineering",
    25: "Engineering Technologies",
    26: "Foreign Languages",
    29: "Family and Consumer Sciences",
    32: "Law",
    33: "English/Literature",
    34: "Liberal Arts and Humanities",
    35: "Library Science",
    36: "Biology and Life Sciences",
    37: "Mathematics and Statistics",
    38: "Military Technologies",
    40: "Interdisciplinary Studies",
    41: "Parks/Recreation/Leisure",
    48: "Philosophy and Religious Studies",
    49: "Theology",
    50: "Physical Sciences",
    51: "Nuclear/Industrial Technology",
    52: "Psychology",
    53: "Criminal Justice",
    54: "Public Affairs and Social Work",
    55: "Social Sciences",
    56: "Construction Services",
    57: "Electrical and Mechanical Repair",
    58: "Precision Production and Industrial Arts",
    59: "Transportation Sciences",
    60: "Fine Arts",
    61: "Medical and Health Sciences",
    62: "Business",
    64: "History"
}

df['Major'] = df["Major"].map(major_map)
```

```python
def collapse_edu(x):
    if x < 62:
        return "HS or Less"
    elif 62 <= x < 101:
        return "Some College"
    elif x == 101:
        return "Bachelor's"
    else:
        return "Masters or Doctorate"

df["Education Group"] = df["Education Level (Detailed)"].apply(collapse_edu)
```

```python
# Map numeric FIPS state codes to state names
state_fips_map = {
    1: "Alabama",
    2: "Alaska",
    4: "Arizona",
    5: "Arkansas",
    6: "California",
    8: "Colorado",
    9: "Connecticut",
    10: "Delaware",
    11: "District of Columbia",
    12: "Florida",
    13: "Georgia",
    15: "Hawaii",
    16: "Idaho",
    17: "Illinois",
    18: "Indiana",
    19: "Iowa",
    20: "Kansas",
    21: "Kentucky",
    22: "Louisiana",
    23: "Maine",
    24: "Maryland",
    25: "Massachusetts",
    26: "Michigan",
    27: "Minnesota",
    28: "Mississippi",
    29: "Missouri",
    30: "Montana",
    31: "Nebraska",
    32: "Nevada",
    33: "New Hampshire",
    34: "New Jersey",
    35: "New Mexico",
    36: "New York",
    37: "North Carolina",
    38: "North Dakota",
    39: "Ohio",
    40: "Oklahoma",
    41: "Oregon",
    42: "Pennsylvania",
    44: "Rhode Island",
    45: "South Carolina",
    46: "South Dakota",
    47: "Tennessee",
    48: "Texas",
    49: "Utah",
    50: "Vermont",
    51: "Virginia",
    53: "Washington",
    54: "West Virginia",
    55: "Wisconsin",
    56: "Wyoming"
}

df["State"] = df["State FIP Code"].map(state_fips_map)
```

# **Convert Nominal Incomes**
---

```python
cpi = pd.read_excel("CPI.xlsx", index_col="Year")
cpi.index = cpi.index.astype(int)
```

```python
base_year = 2024
base_cpi = cpi.loc[base_year, "Annual Average CPI(-U)"]

cpi["adjustment_factor"] = base_cpi / cpi["Annual Average CPI(-U)"]
```

```python
df = df.merge(
    cpi["adjustment_factor"],
    left_on="Year",
    right_index=True,
    how="left"
)
```

```python
df["Income (2024 Dollars)"] = df["Income (Nominal)"] * df["adjustment_factor"]
```

```python
df["Log Income"] = np.log(df["Income (2024 Dollars)"])
```

# **Final Dataframe**
---

```python
df = df.reset_index(drop = True)
```

```python
cat_cols = [
    "State",
    "Sex",
    "Demographic",
    "Education Group",
    "Major",
    "Education Level (Detailed)"
]

for col in cat_cols:
    df[col] = df[col].astype("category")
```

```python
# Memory Optimizations
df["Year"] = df["Year"].astype("int16")
df["Individuals Represented"] = df["Individuals Represented"].astype("int32")
df["Age"] = df["Age"].astype("int8")
df["Hours Worked"] = df["Hours Worked"].astype("int8")
df["Weeks Worked"] = df["Weeks Worked"].astype("int8")
df["Income (Nominal)"] = df["Income (Nominal)"].astype("int32")
df["Income (2024 Dollars)"] = df["Income (2024 Dollars)"].astype("float32")
df["Log Income"] = df["Log Income"].astype("float32")
df["Education Level (Detailed)"] = df["Education Level (Detailed)"].astype("int16")
```

```python
df = df.drop(columns = ['State FIP Code', 'Race', 'Race (Detailed)', 'Hispanic', 'Hispanic (Detailed)', 'adjustment_factor', 'Major (Detailed)', 'Employment Status', 'Employment Status (Detailed)', 'Education Level'])
```
